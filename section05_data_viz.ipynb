# Assignment 3: Subplots

Filter the transactions table to stores 1, 2, 3 and 4, then reshape it so each row is a date, and the columns of the table are the transactions for each store. 

Once you've done that, create a 2x2 line plot of transactions vs time, with one plot for each store.

Thanks!

```python
transactions = pd.read_csv("../../retail/transactions.csv")
transactions
```

```python
pivot_trans = (transactions
               .astype({"date":"datetime64[ns]"})
               .query("store_nbr in [1, 2, 3, 4]")
               .pivot_table(
                   index="date", 
                   columns="store_nbr", 
                   aggfunc={"transactions":"sum"}
               )
               .droplevel(0, axis=1) # remove outer column index level
)
pivot_trans
```

```python
pivot_trans.plot(
    figsize=(14,8),
    xlabel="",
    subplots=True,
    layout=(2, 2),
    sharey=True,
    legend=False,
    title=["Store 1", "Store 2", "Store 3","Store 4"]
)
```

```python
pivot_trans_monthly.sort_index(ascending=False).plot.barh(stacked=True).legend(bbox_to_anchor=(1, 1))
```
